```python
from pyspark.sql import SparkSession

try:
    sc.stop()
except:
    pass

spark = SparkSession.builder \
    .appName("Examples on Map Reduce") \
    .master("spark://4c7cfe53bbf6:7077") \
    .config("spark.ui.port", "4040") \
    .getOrCreate()

sc = spark.sparkContext
sc.setLogLevel("ERROR")
```

```python
temperatures_celsius = [0, 10, 20, 30, 40]
rdd = sc.parallelize(temperatures_celsius)
```

```python
farenheit=rdd.map(lambda x: (x*9/5)+32)
final= farenheit.collect()
```

```python
print("temperatures in farenheit", final) 
```
