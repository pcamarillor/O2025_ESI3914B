# Create SparkSession

```python
import findspark
findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Examples on Structured Streaming (files)") \
    .master("spark://spark-master:7077") \
    .config("spark.ui.port", "4040") \
    .getOrCreate()

sc = spark.sparkContext
sc.setLogLevel("ERROR")
# Optimization (reduce the number of shuffle partitions)
spark.conf.set("spark.sql.shuffle.partitions", "5")
```

# Create a data stream from files

```python
from jaimodule.logGenerator import generate

path = "/opt/spark/work-dir/data/logs"

stream = spark.readStream \
            .format("text") \
            .load(path)

filtered = stream.filter("value LIKE '%Error%' OR value LIKE '%Unauthorized%'")

query = filtered.writeStream \
    .outputMode("append") \
    .format("console") \
    .option("truncate", False)\
    .start()

generate()

query.awaitTermination(60) # Wait 30 seconds before closing the stream
```
